# OUTLIER DETECTION AND REMOVING OF OUTLIERS

```python
def get_numeric_columns(df):
    return df.select_dtypes(include=[np.number]).columns
```

```python
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]
```

```python
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df
```

```python
numeric_columns = get_numeric_columns(df)
```

```python
outliers_summary = {}
for column in numeric_columns:
    outliers = detect_outliers(df, column)
    outliers_count = len(outliers)
    outliers_summary[column] = outliers_count
    if outliers_count > 0:
        print(f"Outliers detected in {column}: {outliers_count}")
        df = handle_outliers(df, column)
```

```python
print("\nSummary of outliers:")
for column, count in outliers_summary.items():
    print(f"{column}: {count}")
```

# FINDING THE BEST PARAMETRES FOR THE REGRESSION MODELS

```python
from sklearn.model_selection import GridSearchCV

def grid_search_tuning(X_train, y_train):
    param_grid = {
        'Lasso': {'alpha': [0.01, 0.1, 1, 10]},
        'Ridge': {'alpha': [0.01, 0.1, 1, 10]},
        'DecisionTreeRegressor': {'max_depth': [3, 5, 7, 10]},
        'KNeighborsRegressor': {'n_neighbors': [3, 5, 7, 9]},  
        'RandomForestRegressor': {'n_estimators': [50, 100, 200], 'max_depth': [5, 10, 15]},
        'GradientBoostingRegressor': {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2], 'max_depth': [3, 5, 7]}} 
    
    for model_name, params in param_grid.items():
        if model_name == 'Lasso':
            model = Lasso()
        elif model_name == 'Ridge':
            model = Ridge()
        elif model_name == 'DecisionTreeRegressor':
            model = DecisionTreeRegressor(random_state=0)
        elif model_name == 'KNeighborsRegressor':
            model = KNeighborsRegressor()
        elif model_name == 'RandomForestRegressor':
            model = RandomForestRegressor(random_state=0)
        elif model_name == 'GradientBoostingRegressor':
            model = GradientBoostingRegressor(random_state=0)
        
        grid = GridSearchCV(model, param_grid=params, cv=5, scoring='r2')
        grid.fit(X_train, y_train)
        print(f"Best parameters for {model_name}: {grid.best_params_}")
        print(f"Best cross-validation R2 score for {model_name}: {grid.best_score_}")
```

```python
grid_search_tuning(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, accuracy_score, mean_squared_error
```

```python
def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    print(f"{model.__class__.__name__} Results")
    print("Accuracy Score at training data:", model.score(X_train, y_train))
    print("Accuracy score at testing data:", model.score(X_test, y_test))
    print("R2 Score:", r2_score(y_test, y_pred))
    mse = mean_squared_error(y_test, y_pred)
    print("Mean Squared Error:", mse)
    print("Root Mean Squared Error:", np.sqrt(mse))
    print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
    print("\n")
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Lasso Regression": Lasso(alpha= 0.01),
    "Ridge Regression": Ridge(alpha= 1),
    "Decision Tree Regression": DecisionTreeRegressor(max_depth = 10, random_state=0),
    "Gradient Boosting": GradientBoostingRegressor(random_state=0, n_estimators = 50, learning_rate = 0.1, max_depth = 3),
    "Random Forest": RandomForestRegressor(random_state=0, n_estimators = 200, max_depth = 3),
    "KNN Regressor": KNeighborsRegressor(n_neighbors = 3), 

}
```

```python
for name, model in models.items():
    print(f"Evaluating {name}")
    evaluate_model(model, X_train, X_test, y_train, y_test)
```

# CHALLENGES FACED


1. Handling Missing and Duplicate Data:

Challenge: The dataset was checked for missing values and duplicates, as these can skew the model’s predictions and reduce its accuracy.

Solution: df.isnull().sum() function was used to identify missing data and df.duplicated().sum() was used to check for duplicates. Since there were no missing or duplicate values reported, no further action was required.

2. Categorical Data Encoding:

Challenge: The dataset contained categorical variables (e.g., CarName, fueltype, carbody, etc.), which need to be converted into a numerical format for machine learning algorithms to process.

Solution: LabelEncoder was used to convert these categorical variables into numerical values. This helped the models to interpret the features correctly.

3. Outliers in Numeric Data:

Challenge: Outliers can negatively affect the performance of regression models by distorting the learning process and making the model biased toward extreme values.

Solution: The IQR (Interquartile Range) method was used to detect and handle outliers in numeric columns. By defining upper and lower bounds based on the IQR, outliers were removed or adjusted.

4. Scaling Features:

Challenge: The dataset contained variables with different ranges, and unscaled data can cause certain algorithms (e.g., KNN, Ridge) to behave poorly.

Solution: Robust Scaler was used to scale the data. This method is less sensitive to outliers compared to other scaling techniques, ensuring that all features were normalized for better model performance.

5. Model Hyperparameter Tuning:

Challenge: Choosing the correct hyperparameters is crucial for optimal model performance, but manually tuning hyperparameters is inefficient.

Solution: GridSearchCV was employed for hyperparameter tuning. It helped in finding the best hyperparameters for several models (e.g., Lasso, Ridge, Decision Tree, KNN, Random Forest, Gradient Boosting), improving the overall accuracy and R² scores of these models.

6. Overfitting and Model Selection:

Challenge: Some models, like Decision Tree and Random Forest, tended to overfit on the training data, leading to high accuracy during training but lower performance on the testing set.

Solution: The usage of GridSearchCV helped optimize model depth and regularization parameters, reducing overfitting. Models like Gradient Boosting Regressor and Stacking Regressor were considered as they performed better in balancing training and testing performance.
