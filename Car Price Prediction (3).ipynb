# OUTLIER DETECTION AND REMOVING OF OUTLIERS

```python
def get_numeric_columns(df):
    return df.select_dtypes(include=[np.number]).columns
```

```python
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]
```

```python
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df
```

```python
numeric_columns = get_numeric_columns(df)
```

```python
outliers_summary = {}
for column in numeric_columns:
    outliers = detect_outliers(df, column)
    outliers_count = len(outliers)
    outliers_summary[column] = outliers_count
    if outliers_count > 0:
        print(f"Outliers detected in {column}: {outliers_count}")
        df = handle_outliers(df, column)
```

```python
print("\nSummary of outliers:")
for column, count in outliers_summary.items():
    print(f"{column}: {count}")
```

# FINDING THE BEST PARAMETRES FOR THE REGRESSION MODELS

```python
from sklearn.model_selection import GridSearchCV

def grid_search_tuning(X_train_scaled, y_train_scaled):
    param_grid = {
        'Lasso': {'alpha': [0.01, 0.1, 1, 10]},
        'Ridge': {'alpha': [0.01, 0.1, 1, 10]},
        'DecisionTreeRegressor': {'max_depth': [3, 5, 7, 10]},
        'KNeighborsRegressor': {'n_neighbors': [3, 5, 7, 9]},  
        'RandomForestRegressor': {'n_estimators': [50, 100, 200], 'max_depth': [5, 10, 15]},
        'GradientBoostingRegressor': {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2], 'max_depth': [3, 5, 7]}} 
    
    for model_name, params in param_grid.items():
        if model_name == 'Lasso':
            model = Lasso()
        elif model_name == 'Ridge':
            model = Ridge()
        elif model_name == 'DecisionTreeRegressor':
            model = DecisionTreeRegressor(random_state=0)
        elif model_name == 'KNeighborsRegressor':
            model = KNeighborsRegressor()
        elif model_name == 'RandomForestRegressor':
            model = RandomForestRegressor(random_state=0)
        elif model_name == 'GradientBoostingRegressor':
            model = GradientBoostingRegressor(random_state=0)
        
        grid = GridSearchCV(model, param_grid=params, cv=5, scoring='r2')
        grid.fit(X_train, y_train)
        print(f"Best parameters for {model_name}: {grid.best_params_}")
        print(f"Best cross-validation R2 score for {model_name}: {grid.best_score_}")
```

```python
grid_search_tuning(X_train_scaled, y_train_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, accuracy_score, mean_squared_error
```

```python
def evaluate_model(model, X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled):
    model.fit(X_train_scaled, y_train_scaled)
    y_pred = model.predict(X_test_scaled)
    
    print(f"{model.__class__.__name__} Results")
    print("Accuracy Score at training data:", model.score(X_train_scaled, y_train_scaled))
    print("Accuracy score at testing data:", model.score(X_test_scaled, y_test_scaled))
    print("R2 Score:", r2_score(y_test_scaled, y_pred))
    mse = mean_squared_error(y_test_scaled, y_pred)
    print("Mean Squared Error:", mse)
    print("Root Mean Squared Error:", np.sqrt(mse))
    print("Mean Absolute Error:", mean_absolute_error(y_test_scaled, y_pred))
    print("\n")
```

```python
import joblib
import numpy as np
import pandas as pd

def predict_car_price(new_data_dict):
    # ✅ Load saved components
    encoder = joblib.load("ordinal_encoder.pkl")
    scaler_X = joblib.load("scaler_X.pkl")
    scaler_y = joblib.load("scaler_y.pkl")
    rf_model = joblib.load("random_forest_model.pkl")
    gb_model = joblib.load("gradient_boost_model.pkl")
    stacking_model_4 = joblib.load("stacking_regressor_model.pkl")
    feature_columns = joblib.load("feature_columns.pkl")

    # ✅ Convert new data to DataFrame
    new_data = pd.DataFrame([new_data_dict])

    # ✅ Step 1: Encode categorical columns safely
    categorical_features = list(encoder.feature_names_in_)

    # Ensure all categorical columns exist in the input
    for col in categorical_features:
        if col not in new_data.columns:
            new_data[col] = "unknown"

    # Replace unseen categories with the first known category from training
    for i, col in enumerate(categorical_features):
        known_categories = encoder.categories_[i]
        new_data[col] = new_data[col].apply(lambda x: x if x in known_categories else known_categories[0])

    # Transform categorical features
    new_data[categorical_features] = encoder.transform(new_data[categorical_features])

    # ✅ Step 2: Align with feature columns
    missing_cols = set(feature_columns) - set(new_data.columns)
    for col in missing_cols:
        new_data[col] = 0
    new_data = new_data[feature_columns]

    # ✅ Step 3: Scale and Predict
    new_data_scaled = scaler_X.transform(new_data)

    rf_pred_scaled = rf_model.predict(new_data_scaled)
    gb_pred_scaled = gb_model.predict(new_data_scaled)
    stack_pred_scaled = stacking_model_4.predict(new_data_scaled)

    rf_pred = scaler_y.inverse_transform(rf_pred_scaled.reshape(-1, 1))[0][0]
    gb_pred = scaler_y.inverse_transform(gb_pred_scaled.reshape(-1, 1))[0][0]
    stack_pred = scaler_y.inverse_transform(stack_pred_scaled.reshape(-1, 1))[0][0]

    # ✅ Step 4: Combine predictions
    avg_pred = np.mean([rf_pred, gb_pred, stack_pred])
    weighted_pred = (rf_pred * 0.4) + (gb_pred * 0.3) + (stack_pred * 0.3)

    # ✅ Step 5: Print & Return
    print("Random Forest Prediction:", round(rf_pred, 2))
    print("Gradient Boost Prediction:", round(gb_pred, 2))
    print("Stacking Regressor Prediction:", round(stack_pred, 2))
    print("Average Combined Prediction:", round(avg_pred, 2))
    print("Weighted Combined Prediction:", round(weighted_pred, 2))

    return {
        "RandomForest": rf_pred,
        "GradientBoost": gb_pred,
        "Stacking": stack_pred,
        "Average": avg_pred,
        "Weighted": weighted_pred
    }
```
